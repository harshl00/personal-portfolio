```python
import pandas as pd
data = pd.read_csv("C:/Users/My PC/Downloads/archive/nyc_weather.csv")
df=data.sort_values(by=['Temperature'])
df.head()
```

```python
#compare with yesterday
data['Temperature_lag']=data["Temperature"].shift(1)
data.head()
```

```python
#compare with nextday
data['Temperature_lead']=data["Temperature"].shift(-1)
data.head()
```

```python
#rolling average for 5 day moving average
data["Temperature_avg"]=data["Temperature"].rolling(5).mean()
data.head()
```

```python
#rolling sum for 2 days
data["temp_sum"]=data.Temperature.rolling(2).sum()
data["temp_sum1"]=data.Temperature.expanding(2).sum()
data.head()
```

```python
#for 5 max and min
data["temp_max"]=data["Temperature"].rolling(7).max()
data["temp_min"]=data["Temperature"].rolling(7).min()
data.head(7)
```

```python
data['temp_pctchange']=data.Temperature.pct_change(1)*100
data.head()
```
